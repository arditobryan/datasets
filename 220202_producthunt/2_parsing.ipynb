```python
import requests
from bs4 import BeautifulSoup
import os

li_list = list()
for page in os.listdir()[2:]:
    if page[-3:] == 'txt':
        print(page)
        with open(page) as file:
            text = file.read()

        soup = BeautifulSoup(text, 'html.parser')
        soup

        li = soup.find_all('li')
        li

        def extract_values(html):
            #print(html)
            dict1 = dict()
            dict1['title'] = html.find_all('a')[1].decode_contents()
            dict1['type'] = page[:-3]
            dict1['description'] = html.find_all('a')[3].decode_contents()
            dict1['upvotes'] = html.find_all('a')[4].find('div').find('div').decode_contents()
            dict1['reviews'] = html.find_all('a')[5].find('div').decode_contents()

            try:
                #img
                dict1['img'] = html.find_all('a')[0].find_all('div')[0].find('div').find('video').__dict__['attrs']['poster'].split('?')[0]
            except:
                #gif
                dict1['img'] = html.find_all('a')[0].find_all('div')[0].find('div').find('img').__dict__['attrs']['srcset'].split('?')[0]
            #print(dict1)
            return dict1
        
        for el_index in range(5, len(li)): #len(li)
            try:
                dict1 = extract_values(li[el_index])
                li_list.append(dict1)
            except:
                #print(el_index)
                pass
```

```python
import pandas as pd

df = pd.DataFrame(li_list)
df['type'] = df['type'].apply(lambda x : x[:-1])
df['upvotes'] = df['upvotes'].apply(lambda x : x[:-1])
df['reviews'] = df['reviews'].apply(lambda x : int(x.split(' ')[0]))
df
```

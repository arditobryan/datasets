```python
import pandas as pd
import numpy as np

df = pd.read_csv('3_mpst_drop_encoded.csv') 
df.index = df.pop('title')
df
```

```python
for col in df.columns:
    col = str(col)
    df[col] = np.where(df[col]<=0, 0, 1)
```

```python
final = df.groupby(list(df.columns), group_keys=True, squeeze=True, sort=True)

#
display(final.count())

#
df_grouped = pd.DataFrame(final.size().sort_values(ascending=False))
display(df_grouped)
```

```python
#export
df_grouped.to_csv('4_mpst_drop_binary.csv')
```

```python
#export
df_grouped = pd.read_csv('4_mpst_drop_binary.csv')
df_grouped
```

```python
#   I am extracting the first pseudocluster with 12 movies together
#df_grouped is the final that contains the frequency
#df is contains the original encoded values
pseudocluster_rows = list()
for row in range(len(df)):
    val = df.iloc[row].values

    if np.all(val == df_grouped.iloc[0].values[:-1]) == True:
        pseudocluster_rows.append(row)
```

```python
from bs4 import BeautifulSoup
import requests
import time
import random
import pandas as pd

list_samples = list()
for page_n in range(1, 34):
    time.sleep(10+random.randint(2, 10))
    print(page_n)

    try:
        soup = BeautifulSoup(requests.get(f'https://patents.justia.com/assignee/uber-technologies-inc?page={page_n}').text)
        soup

        for sample in soup.find_all('li', {'class':'has-padding-content-block-30 -zb'}):
            dict1 = dict()
            dict1['title'] = sample.find('h6', {'class':'heading-6 has-no-margin'}).find('a').decode_contents()
            dict1['date_of_patent'] = sample.find('div', {'class':'date-filed'}).decode_contents().split('\n')[2].strip()
            dict1['date_of_patent'] = sample.find('div', {'class':'date-issued'}).decode_contents().split('\n')[2].strip()
            dict1['patent_n'] = sample.find('div', {'class':'number'}).decode_contents().split(' ')[3][:-2]
            dict1['description'] = sample.find('div', {'class':'abstract'}).decode_contents()[29:]
            dict1['type'] = sample.find('div', {'class':'type'}).decode_contents().split('\n')[2].replace(' ','')
            dict1['assignees'] = sample.find('div', {'class':'assignees'}).decode_contents().split('\n')[2].replace(' ','')
            dict1['inventors'] = sample.find('div', {'class':'inventors'}).decode_contents().split('\n')[3].strip()
            #dict1['inventors'] = sample.find('div', {'class':'inventors'}).decode_contents().split('\n')[3].strip().split(',')
            list_samples.append(dict1)
    except:
        pass

df = pd.DataFrame(list_samples)
df.to_csv('patents_uber_2.csv', index=None)
df
```

## groupby_frequency

```python
import pandas as pd
import numpy as np

df_matches = pd.read_csv('CRL_2_matches.csv').drop('Unnamed: 0', axis=1)
df_onehot = pd.read_csv('CRL_3_encoded.csv').drop('Unnamed: 0', axis=1)
df_onehot
```

```python
grouped = df_onehot.groupby(list(df_onehot.columns), group_keys=True, squeeze=True, sort=True)

#grouped
df_grouped = grouped.count()
display(df_grouped)

#grouped+frequency
df_grouped_freq = pd.DataFrame(grouped.size().sort_values(ascending=False))
display(df_grouped_freq)
```

```python
#export
df_grouped_freq.to_csv('CRL_5_grouped_freq.csv')
```

## pseudoclusters

```python
#import
df_grouped_freq = pd.read_csv('CRL_5_grouped_freq.csv')
#df_grouped_freq.iloc[0]
```

```python
#first pseudocluster
pseudocluster_index = 1
pseudocluster_rows = list()
for row in range(len(df_grouped_freq)):
    val = df_onehot.iloc[row].values
    if np.all(val == df_grouped_freq.iloc[pseudocluster_index].values[:-1]) == True:
        pseudocluster_rows.append(row)
print(len(pseudocluster_rows))
```

```python
display(df_matches.iloc[pseudocluster_rows])
df_matches.iloc[pseudocluster_rows].result.value_counts()
```

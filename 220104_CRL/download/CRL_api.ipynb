```python
import re

tag_list = list()
for tag in [m.start() for m in re.finditer('"tag": ', str1)]:
    str_piece = str1[tag+8:tag+100]
    str_tag = str_piece[0:str_piece.find('"')+2]
    if "}" not in str_tag[-2:]:
        tag_list.append(str_tag[0:-2])
```

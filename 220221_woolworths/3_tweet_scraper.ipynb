```python
#importing libraries and packages
import snscrape.modules.twitter as sntwitter
import pandas as pd

#creating list to append tweet data 
tweets_list1 = []

#using TwitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterUserScraper('woolworths').get_items()): #declare a username 
    #print(tweet.content)
    if i>1200: #number of tweets you want to scrape
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username]) #declare the attributes to be returned
    
#creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
tweets_df1
```

```python
def count_consecutives(list1):
    final = list()
    counter = 1
    for index in range(len(list1)-1):
        #print(index, list1[index])
        if list1[index] == list1[index+1]:
            final.append(counter)
            counter += 1
        else:
            final.append(counter)
            counter = 1

    #solve for the last appearence  
    if list1[-1] == list1[-2]:
        final.append(final[-1]+1)
    else:
        final.append(1)

    final = [str(x) for x in final]
    return final

def user_id(tweet):
    for word in tweet.split(' '):
        if word.find('@') != -1:
            user = word
            break
    return user

global c 
c = 0

def search_response(user_):
    global c
    c+=1
    print(c)
    user = user_.split(':')[0]
    counter = int(user_.split(':')[1])
    n_found = 0
    print(user, counter)
    #using TwitterSearchScraper to scrape data and append tweets to list
    for i, tweet in enumerate(sntwitter.TwitterUserScraper(user[1:]).get_items()): #declare a username 
        #print('B', user, i)
        if tweet.content.find('@woolworths') != -1:
            n_found += 1
            #print(tweet.content)
            #print('n_found', n_found)
            #print('counter', counter)
            #reply found
            if counter == n_found:
                #print('RETURNING')
                return tweet.content
        if i > 220:
            break

df = tweets_df1[['Text']]
df['User'] = df['Text'].apply(lambda x : user_id(x))
df = df.sort_values('User')
df['Counter'] = count_consecutives(df['User'].values)
df['User'] = df['User'] + ':'+ df['Counter']
df['Reply'] = df['User'].apply(lambda x : search_response(x))
df
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import os

sample_list = list()
for page in os.listdir('files'):
    if page.find('Lyft') != -1:
        soup = BeautifulSoup(open(f'files/{page}').read())

        for el in soup.find_all('div', {'class':'p-0 mb-0 mb-md-std css-w5wad1 gd-ui-module css-rntt2a ec4dwm00'})[0:10]:
            dict1 = dict()
            dict1['rating'] = el.find('span').decode_contents()
            dict1['review'] = el.find('a', {'class':'reviewLink'}).decode_contents()

            if el.find_all('svg')[0].find('cicle') == None:
                #path found
                dict1['recommended'] = True
            else:
                #circle found
                dict1['recommended'] = False
            dict1['pros'] = el.find('span', {'data-test':'pros'}).decode_contents()
            dict1['cons'] = el.find('span', {'data-test':'cons'}).decode_contents()
            sample_list.append(dict1)

df = pd.DataFrame(sample_list)
df['rating'] = df['rating'].apply(lambda x: int(x[0]))
df
```

```python
df.to_csv('Lyft_reviews.csv', index=None)
```

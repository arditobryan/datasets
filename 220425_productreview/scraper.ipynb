```python
import requests
from bs4 import BeautifulSoup
import time
import random

sample_list = list()
type_list = [
  'products',
 'appliances',
 'electronics',
 'beauty-products',
 'babies-kids',
 'furniture-bedding',
 'garden-power-tools',
 'health-products',
 'food-drink',
 'vehicles',
 'home-care',
 'travel-products',
 'automotive-accessories',
 'personal-care-hygiene',
 'pets-products',
 'sport-recreation',
 'kitchenware',
 'watches',
 'rapid-antigen-self-tests',
 'services',
]

for type_ in type_list:
  for page_n in range(0, 12):
    time.sleep(30+random.randint(2, 10))

    URL = f'https://www.productreview.com.au/c/{type_}?page={page_n}#search-results'
    page = requests.get(URL)

    #next step: parsing
    soup = BeautifulSoup(page.content, 'html.parser')
    #soup

    if len(soup.find_all('div', {'class':'col-md-9__CL col-lg-12_H3a col-xl-9__Db flex-column__BR relative___C d-flex_RNf stackY-3_Tar'})) == 0: print('empty')
    for el in soup.find_all('div', {'class':'col-md-9__CL col-lg-12_H3a col-xl-9__Db flex-column__BR relative___C d-flex_RNf stackY-3_Tar'}):
      dict1 = dict()
      try:
          dict1['type'] = type_
      except:
          print('type', el)

      try:
        dict1['description'] = el.find_all('span', {'class':'font-size-sm_C3A trim-y_Kjv d-block__Gv'})[0].decode_contents()[3:-5]
      except:
        try:
          dict1['description'] = el.find_all('span', {'class':'text-break_e5Q text-italic_e_V'})[0].decode_contents()[3:-5]
        except:
          print(el, 'description')

      try:
        dict1['title'] = el.find_all('a', {'class':'text-link_VhC stretched-link_DLa text-break_e5Q text-decoration-underline-hover_K2N'})[0].decode_contents()
      except:
        print(el, 'title')

      try:
        for img in soup.find_all('img'):
          if img.get('alt') == dict1['title']:
            dict1['img'] = img.get('src')
            break

        for img in soup.find_all('img'):
          if img.get('alt') == dict1['title'].split(' (')[0]:
            dict1['img'] = img.get('src')
            break
      
      except:
          print('img', el)

      try:
        dict1['score'] = el.find_all('div', {'class':'align-items-center_amm flex-wrap_tFJ d-flex_RNf'})[0].find_all('span')[1].decode_contents()
      except:
        print(el, 'score')

      try:
        dict1['review_n'] = el.find_all('div', {'class':'align-items-center_amm flex-wrap_tFJ d-flex_RNf'})[0].find_all('span')[2].decode_contents().split('<')[0]
      except:
        print(el, 'review_n')
      sample_list.append(dict1)

    print(type_, page_n, len(sample_list))
sample_list

import pandas as pd
df = pd.DataFrame(sample_list)
df.to_csv('t3.csv', index=None)
```

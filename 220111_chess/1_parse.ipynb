```python
import pandas as pd

df = pd.read_csv('lichess_1_raw.csv').drop('Unnamed: 0', axis=1)
df = df.dropna()
df
```

```python
df_game = df[['PGN']][0:25000]

#split into lists
df_game['PGN'] = df_game['PGN'].apply(lambda x: x.split(' '))

#delete numbers
df_game['PGN'] = df_game['PGN'].apply(lambda x: [a for a in x if (a.find('.') == -1)])

#
#df_game['PGN'] = df_game['PGN'].apply(lambda x: x[0:5])
df_game['PGN']
```

```python
import numpy as np

#NON FUNZIONA SEMPRE
pd.DataFrame(np.column_stack(list(zip(*df_game.values))))

#funziona, elementi mancanti diventano None
df_game = pd.DataFrame([x[0] for x in df_game.values])
df_game
```

## encoding

```python
from sklearn import preprocessing
codex = preprocessing.LabelEncoder()
codex = codex.fit(pd.unique(df_game.values.ravel('K'))) #all unique values in a dataset
#codex.transform(['Battle Healer'])
#codex.inverse_transform([10])

df_ready_pd = df_game.apply(lambda x : codex.transform(x))
df_ready_pd = pd.DataFrame([[list(df_ready_pd.iloc[x].values)] for x in range(len(df_ready_pd))])
df_ready_pd = df_ready_pd.reset_index()
df_ready_pd.columns = ['_id', 'match_vector_']
df_ready_pd['match_vector_'] = df_ready_pd['match_vector_'].apply(lambda x: [float(a) for a in x])
df_ready_pd['original'] = pd.DataFrame([[list(df_game.iloc[x].values)] for x in range(len(df_game))])[0]
df_ready_pd
```

```python
import json

df_ready = list()
for a in range(int(len(df_ready_pd)/50)):
    [df_ready.append(x) for x in df_ready_pd[a*50:(a+1)*50].to_dict(orient='records')]
df_ready
```

```python
import json

#export to json
with open('lichess_2_ready.json', 'w') as f:
    json.dump(df_ready, f)
```
